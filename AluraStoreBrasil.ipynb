### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
loja3.head()
```

```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()
```

```python
print(f"Faturamento total da loja 1: R$ {faturamento_loja1}")
print(f"Faturamento total da loja 2: R$ {faturamento_loja2}")
print(f"Faturamento total da loja 3: R$ {faturamento_loja3}")
print(f"Faturamento total da loja 4: R$ {faturamento_loja4}")
```

# 2. Vendas por Categoria

```python
#venda por categoria da loja 1:
venda_categora_l1 = loja.groupby('Categoria do Produto')['Categoria do Produto'].count().sort_values(ascending=False)
venda_categora_l1.head()
```

```python
from numpy._core.defchararray import count
#venda por categoria da loja 2:
venda_categora_l2 = loja2['Categoria do Produto'].value_counts().sort_values(ascending=False)
venda_categora_l2.head()
```

```python
#venda por categoria da loja 3:
venda_categora_l3 = loja3.groupby('Categoria do Produto')['Categoria do Produto'].count().sort_values(ascending=False)
venda_categora_l3.head()
```

```python
#venda por categoria da loja 4:
venda_categora_l4 = loja4.groupby('Categoria do Produto')['Categoria do Produto'].count().sort_values(ascending=False)
venda_categora_l4.head()
```

# 3. Média de Avaliação das Lojas


```python
media_avaliacao_l1 = loja['Avaliação da compra'].mean()
media_avaliacao_l2 = loja2['Avaliação da compra'].mean()
media_avaliacao_l3 = loja3['Avaliação da compra'].mean()
media_avaliacao_l4 = loja4['Avaliação da compra'].mean()

print(f"Média de avaliação da Loja 1: {media_avaliacao_l1:.2f}")
print(f"Média de avaliação da Loja 2: {media_avaliacao_l2:.2f}")
print(f"Média de avaliação da Loja 3: {media_avaliacao_l3:.2f}")
print(f"Média de avaliação da Loja 4: {media_avaliacao_l4:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
# ===== Loja 1 =====
produtos_l1 = loja['Produto'].value_counts()
mais_vendido_l1 = produtos_l1.idxmax()
menos_vendido_l1 = produtos_l1.idxmin()

# ===== Loja 2 =====
produtos_l2 = loja2['Produto'].value_counts()
mais_vendido_l2 = produtos_l2.idxmax()
menos_vendido_l2 = produtos_l2.idxmin()

# ===== Loja 3 =====
produtos_l3 = loja3['Produto'].value_counts()
mais_vendido_l3 = produtos_l3.idxmax()
menos_vendido_l3 = produtos_l3.idxmin()

# ===== Loja 4 =====
produtos_l4 = loja4['Produto'].value_counts()
mais_vendido_l4 = produtos_l4.idxmax()
menos_vendido_l4 = produtos_l4.idxmin()

print("Loja 1")
print(f"Produto mais vendido (+): {mais_vendido_l1}")
print(f"Produto menos vendido (−): {menos_vendido_l1}")
print("-" * 40)

print("Loja 2")
print(f"Produto mais vendido (+): {mais_vendido_l2}")
print(f"Produto menos vendido (−): {menos_vendido_l2}")
print("-" * 40)

print("Loja 3")
print(f"Produto mais vendido (+): {mais_vendido_l3}")
print(f"Produto menos vendido (−): {menos_vendido_l3}")
print("-" * 40)

print("Loja 4")
print(f"Produto mais vendido (+): {mais_vendido_l4}")
print(f"Produto menos vendido (−): {menos_vendido_l4}")
```

# 5. Frete Médio por Loja

```python
# ===== Loja 1 =====
frete_medio_l1 = loja['Frete'].mean()

# ===== Loja 2 =====
frete_medio_l2 = loja2['Frete'].mean()

# ===== Loja 3 =====
frete_medio_l3 = loja3['Frete'].mean()

# ===== Loja 4 =====
frete_medio_l4 = loja4['Frete'].mean()

print("Frete Médio por Loja")

print(f"Loja 1: R$ {frete_medio_l1:.2f}")
print(f"Loja 2: R$ {frete_medio_l2:.2f}")
print(f"Loja 3: R$ {frete_medio_l3:.2f}")
print(f"Loja 4: R$ {frete_medio_l4:.2f}")
```

Gerando Gráficos com Matplotlib

```python
import matplotlib.pyplot as plt
```

Gráfico 1 — Barras

Faturamento Total por Loja

```python
import matplotlib.pyplot as plt

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [
    faturamento_loja1,
    faturamento_loja2,
    faturamento_loja3,
    faturamento_loja4
]

plt.figure()
plt.bar(lojas, faturamentos)
plt.title('Faturamento Total por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.show()
```

Gráfico 2 — Linha

Frete Médio por Loja

```python
fretes_medios = [
    frete_medio_l1,
    frete_medio_l2,
    frete_medio_l3,
    frete_medio_l4
]

plt.figure()
plt.plot(lojas, fretes_medios, marker='o')
plt.title('Frete Médio por Loja')
plt.xlabel('Lojas')
plt.ylabel('Frete Médio (R$)')
plt.show()
```

Gráfico 3 — Pizza (Setores)

Distribuição da Média de Avaliação por Loja

```python
avaliacoes = [
    media_avaliacao_l1,
    media_avaliacao_l2,
    media_avaliacao_l3,
    media_avaliacao_l4
]

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure()
plt.pie(avaliacoes, labels=lojas, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição da Média de Avaliação por Loja')
plt.show()
```

Relatório Final – Análise das Lojas

Este relatório tem como objetivo analisar o desempenho de quatro lojas pertencentes ao Senhor João, com base em dados de faturamento, categorias de produtos, avaliações dos clientes, produtos mais e menos vendidos e custo médio de frete. A partir dessas análises e das visualizações geradas, busca-se identificar qual loja apresenta o menor desempenho geral e, portanto, deve ser recomendada para venda, apoiando a tomada de decisão estratégica.

A análise do faturamento total mostrou que a Loja 4 apresentou o menor faturamento, totalizando R$ 1.384.497,58, enquanto as demais lojas alcançaram valores superiores. Esse resultado indica menor desempenho financeiro e menor capacidade de geração de receita quando comparada às outras unidades.

Em relação às categorias de produtos mais e menos vendidas, observou-se que a Loja 4 não se destacou positivamente nas categorias com maior volume de vendas, apresentando menor competitividade comercial em relação às demais lojas analisadas.

A média de avaliação dos clientes reforça esse cenário. Embora a Loja 4 apresente uma avaliação média de 4,00, esse valor não representa o melhor desempenho entre as lojas, ficando abaixo da Loja 2 (4,04) e da Loja 3 (4,05). A Loja 1, apesar de apresentar a menor média (3,98), compensa esse resultado com um faturamento significativamente superior.

Quanto aos produtos mais e menos vendidos, a Loja 4 apresenta itens com menor rotatividade, o que pode resultar em estoque parado e menor eficiência operacional, impactando negativamente seus resultados ao longo do tempo.

Por fim, a análise do frete médio, que é pago pelo cliente, mostrou que a Loja 4 possui o menor valor médio (R$ 31,28). No entanto, esse ponto positivo não foi suficiente para impulsionar suas vendas ou compensar o baixo faturamento e o desempenho inferior nos demais indicadores analisados.

Os gráficos gerados (barras, linha e pizza) reforçam visualmente essas conclusões, permitindo uma comparação clara entre as lojas e evidenciando os pontos fortes e fracos de cada uma.

Com base nas análises realizadas, recomenda-se que o Senhor João venda a Loja 4. Essa decisão é justificada pelo fato de a loja apresentar o menor faturamento total, não se destacar nas avaliações dos clientes e não possuir diferenciais competitivos suficientes para compensar seu desempenho inferior, mesmo com um custo médio de frete mais baixo.

As demais lojas demonstram maior potencial financeiro, melhor aceitação dos clientes ou equilíbrio mais favorável entre faturamento e custos, indicando maior sustentabilidade e perspectivas de crescimento. Dessa forma, a venda da Loja 4 representa a escolha mais estratégica e fundamentada nos dados analisados.
